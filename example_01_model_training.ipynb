## ***Training the classification model***

#### *Train the classifier with benchmark-A and RandomForest*

```python
from classifier import RandomForestModel
from preprocess import load_benchmark_dataset
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# Load benchmark dataset
train_dataset_path = "__benchmark_dataset/benchmark_A_train_sequences.csv"
test_dataset_path = "__benchmark_dataset/benchmark_A_test_sequences.csv"
trainset = load_benchmark_dataset(train_dataset_path)
testset = load_benchmark_dataset(test_dataset_path)
dataset = (trainset, testset)

# Initialize classlfier
classifier = RandomForestModel(tag="TESTING-A", num_tree_range=range(35,200))

# Train
logs, model_paths = classifier.train(dataset, num_models=2000, is_save=True)
logs
```
